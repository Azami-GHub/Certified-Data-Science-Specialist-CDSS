 Compute model accuray using the classification algorithm

```python
#step 1 load data and define target and input
from sklearn import datasets
iris = datasets.load_iris()

X = iris.data  #input (#sepal length, sepal width, petal length, petal width)
y = iris.target #target (#species)
```

```python
#step 2 Split and Randomize Data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                test_size=0.3,  random_state = 15)

```

```python
#model evaluation using KNN (k-nearest neighbors)
#Step 3 Define Classifier
from sklearn import neighbors
knn = neighbors.KNeighborsClassifier()
```

```python
#Step 4 Train the Classifier  (Training)
knn.fit(X_train,y_train) #fitting the data into knn algorithm, and it will predict the logic behind input and target
```

```python
#Step 5:	Evaluate the Classifier/ compute accuracy
knn.score(X_test,y_test)
```

```python
#SVM (support vector machine)- classification
from sklearn import svm
svm = svm.SVC()
```

```python
svm.fit(X_train,y_train) #fitting the data into knn algorithm, and it will predict the logic behind input and target
```

```python
svm.score(X_test,y_test)
```

```python
orange = orange.dropna()
```

```python
orange.info()
```

```python
X = orange.drop("quality", axis=1) #select rank and sugar as inputs
y = orange["quality"] #select the target variable
```

```python
#Split and Randomize Data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                test_size=0.3,  random_state = 15)

```

```python
from sklearn import neighbors
knn = neighbors.KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
knn.score(X_test, y_test)
```

```python
#SVM (support vector machine)- classification
from sklearn import svm
svm = svm.SVC()
```

```python
svm.fit(X_train,y_train)
svm.score(X_test, y_test)
```

```python
#Linear Regression - regression
from sklearn import linear_model
lm = linear_model.SGDClassifier()
```

```python
lm.fit(X_train,y_train)
lm.score(X_test, y_test)
```

```python
#Decision tree - classification
from sklearn import tree
dt = tree.DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
dt.score(X_test, y_test)
```
