#***Matplotlib Pyplot***

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias.

The *plot()* function is used to draw points (markers) in a diagram.

By default, the *plot()* function draws a line from point to point.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np #handle numbers/ numerical variables (numerical python)
```

```python
#Draw a line in a diagram from position (0,0) to position (6,250):
xpoints = np.array([0, 6]) #xvariable
ypoints = np.array([0, 250]) #yvariable
plt.plot(xpoints, ypoints) #plot

plt.show() #display
```

plot and show functions are important to visualise graphs in Python

#***Line style***

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

You can choose any of these styles:

* 'solid' (default)    '-'
* 'dotted' ':'
* 'dashed' '--'
* 'dashdot' '-.'
* 'None' '' or ' '

#***Labels and Title***

With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

#***Grid lines***

With Pyplot, you can use the grid() function to add grid lines to the plot.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.grid()
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```

#***Multiple Plots***

With the subplots() function you can draw multiple plots in one figure:

```python
#Example: draw 2 plots

#plot 1:
x = np.array([0, 1, 2, 3]) #array is to create a series of number
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) #row, column, position
plt.plot(x,y)


#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])


plt.subplot(1, 2, 2) #dimension
plt.plot(x,y)



plt.show()
```

# Try to change ***plt.subplot(1, 2, 2)*** to ***plt.subplot(2, 1, 2)***

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])


plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

#**Bar plot**

With Pyplot, you can use the bar() function to draw bar graphs:

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([90, 70, 60, 40])

plt.bar(x,y)
#plt.show()
```

```python
#Horizontal bars
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
#Change bar colors
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red")
plt.show()
```

```python
#change bar width
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
```

```python
#change bar height
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```

#**Pie Charts**

With Pyplot, you can use the pie() function to draw pie charts:

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

```python
#labels
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels) #labels is the default parameter
plt.show()
```

```python
#legends with header
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```
